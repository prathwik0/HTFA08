```python
import requests
```

```python
url = "http://127.0.0.1:5000"
url = "https://trend-predictor-mwnxl4smaa-el.a.run.app"

file1_path = "./BreastCancer.csv"
file2_path = "./AnonymizedBreastCancer.csv"

data = {"target": "diagnosis"}
```

```python
from flask import jsonify


with open(file1_path, "rb") as file1, open(file2_path, "rb") as file2:
    files = {"original": file1, "anonymous": file2}

    response = requests.post(url, files=files, data=data)

    if response.status_code == 200:
        print("Files uploaded successfully.")
        print(response.json())
    else:
        print("Error uploading files. Status code:", response.status_code)
        print(jsonify(response))
```
